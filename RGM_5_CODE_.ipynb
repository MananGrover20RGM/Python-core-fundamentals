## Question 16: Tracking Fruit Stock using Python Sets

**Problem:** A storekeeper wants to compare the fruit varieties in two baskets. We need to find:
* Total unique fruits available across both baskets.
* Fruits available in both baskets (common items).
* Fruits that are completely unique to each basket.

## Question 21: Creating a Shipping Cost Calculator using If-Elif-Else

**Problem:** We need to calculate the cost of shipping a package based on its destination and weight.
* **Domestic Shipping:** Up to 1 kg costs Rs. 50. Each extra kg costs Rs. 10.
* **International Shipping:** Up to 1 kg costs Rs. 500. Each extra kg costs Rs. 100.**bold text**

```python
# --- SOLUTION ---

# Ask the user for the destination type and the package weight
shipping_type = input("Enter shipping type (Domestic / International): ").lower()
weight = float(input("Enter package weight in kg: "))

# Start the cost counter at 0
final_cost = 0

# Check the rules for Domestic shipping
if shipping_type == "domestic":
    if weight <= 1:
        final_cost = 50
    else:
        # Rs. 50 for the first kg + Rs. 10 for each extra kg
        final_cost = 50 + (weight - 1) * 10

# Check the rules for International shipping
elif shipping_type == "international":
    if weight <= 1:
        final_cost = 500
    else:
        # Rs. 500 for the first kg + Rs. 100 for each extra kg
        final_cost = 500 + (weight - 1) * 100

# Print the calculated shipping amount
print("Total Shipping Cost: Rs.", final_cost)
```

## Question 23: Board Game Dice Rolls Average using a For Loop

## Question 23b: Sorting Names by First Letter Using Nested Loops

**Problem:** Given a list of participant names, we need to analyze the starting character of each name and group them into two separate lists: one for names starting with a vowel, and another for consonants.

```python
# --- SOLUTION ---

# Starting list of names provided in the prompt
names = ["Alice", "Bob", "Eve", "Charlie", "Ivy", "David", "Olivia", "Peter"]

# Create empty target lists to store our sorted groups
vowel_names = []
consonant_names = []

# Define a reference string containing all vowels (in lowercase for easy checking)
vowels = "aeiou"

# Loop through each name in our list
for name in names:
    # Grab the very first letter of the name and convert it to lowercase
    first_letter = name[0].lower()

    # Check if that first letter is found inside our vowels string
    if first_letter in vowels:
        vowel_names.append(name)
    else:
        consonant_names.append(name)

# Display both completed lists
print("Names starting with a Vowel:", vowel_names)
print("Names starting with a Consonant:", consonant_names)
```

## Question 25: Counting Even and Odd Numbers in Matrix Lists

**Problem:** We have a nested list of integers. We need to use nested loops to evaluate every individual number and find the total counts of even numbers and odd numbers.

```python
# --- SOLUTION ---

# This is our nested list containing multiple rows of numbers
numbers = [
    [2, 5, 11, 20, 8],
    [9, 4, 15, 28, 17],
    [1, 6, 21, 18, 3],
    [10, 13, 25, 33, 30],
    [14, 7, 16, 19, 22]
]

# Initialize our counting variables to zero before the loop starts
even_count = 0
odd_count = 0

# Outer loop goes through each sub-list row by row
for sub_list in numbers:

    # Inner loop runs through every individual item inside the current sub-list
    for num in sub_list:

        # Check if the number is perfectly divisible by 2 with no remainder
        if num % 2 == 0:
            even_count = even_count + 1
        else:
            odd_count = odd_count + 1

# Display the final counts on the screen
print("Total number of even numbers:", even_count)
print("Total number of odd numbers:", odd_count)
```

* Evaluates user guesses to provide helpful high/low hints until they win or run out of turns.

```python
# --- SOLUTION ---

# 1. Import the "random" module to generate a random number for the game
import random

# Generate a random number between 1 and 100
secret_number = random.randint(1, 100)

# 2. Set the total number of tries allowed and a counter variable to track guesses
max_tries = 7
tries_left = max_tries

# Track if the player guessed successfully
guessed_correctly = False

# 3. Use a while loop to keep guessing until tries run out or they win
while tries_left > 0:
    print(f"\nYou have {tries_left} attempts remaining.")

    # Take input from the player
    guess = int(input("Guess a number between 1 and 100: "))

    # Check if the guess matches the secret number
    if guess == secret_number:
        print("Congratulations! You guessed the correct number!")
        guessed_correctly = True
        break  # Exit the loop instantly since they won
    elif guess < secret_number:
        print("Your guess is too low.")
    else:
        print("Your guess is too high.")

    # Subtract 1 try from our total attempts counter
    tries_left = tries_left - 1

# 4. If they run out of tries without guessing right, reveal the secret answer
if not guessed_correctly:
    print(f"\nGame Over! You've run out of tries. The secret number was {secret_number}.")
```

## Question 35: Processing Numbers Using Lambda Functions

**Problem:** Filter a numerical range from 1 to 10 to keep only odd integers, square them, and find their total sum using `filter()`, `map()`, and `reduce()`.

```python
# --- SOLUTION ---

# Import reduce from the functools library as instructed
from functools import reduce

# Create a list of numbers from 1 to 10
numbers = list(range(1, 11))

# 1. Use filter with a lambda function to keep only odd numbers
odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))

# 2. Use map with a lambda function to square all the filtered odd numbers
squared_odds = list(map(lambda x: x ** 2, odd_numbers))

# 3. Use reduce with a lambda function to add all the squared values together
total_sum = reduce(lambda x, y: x + y, squared_odds)

# Print the final result clearly
print("The sum of squared odd numbers is:", total_sum)
```

## Question : Defining a Time Formatting and Analytics Class

**Problem:** Build a comprehensive `Time` tracking class configuration featuring explicit 24-hour boundary input validation checks, AM/PM calculation layouts, and a zero-safe minutes-to-hours ratio analyzer.

```python
# --- CLASS DEFINITION ---

class Time:
    # The constructor method initializes internal hours and minutes values
    def __init__(self, hours, minutes):
        self.hours = hours
        self.minutes = minutes

    def DisplayTime(self):
        # 1. First validate parameter entry values to ensure safety constraints
        if self.hours >= 24:
            print("The input hours should be less than 24")
            return
        if self.minutes >= 60:
            print("The input minutes should be less than 60")
            return

        # 2. Convert standard 24-hour integers to elegant 12-hour tracking components
        period = "AM"
        display_hour = self.hours

        if self.hours == 0:
            display_hour = 12
        elif self.hours == 12:
            period = "PM"
        elif self.hours > 12:
            display_hour = self.hours - 12
            period = "PM"

        # 3. Format minutes text strings to ensure trailing zeros are preserved
        minute_string = str(self.minutes).zfill(2)
        print(f"The time is {display_hour}:{minute_string} {period}")

    def DisplayRatio(self):
        # Enclose the ratio division within a defensive try block to block crashes
        try:
            ratio_value = self.minutes / self.hours
            print(f"The ratio of minutes to hours is: {ratio_value:.2f}")
        except ZeroDivisionError:
            print("Cannot calculate ratio: Hours value is zero.")
```
